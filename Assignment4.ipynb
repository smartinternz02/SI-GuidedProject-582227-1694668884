```python
#X and y split and Scaling X
x=df.drop(columns=['quality'])
x.head()
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
x=pd.DataFrame(scale.fit_transform(x),columns=x.columns)
```

```python
y=df.quality
y.head()
```

```python
#Splitting in Train and Test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape,x_test.shape
```

Model Building

```python
#Model building
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
r_y_predict=model.predict(x_test)
r_y_predict
r_y_train=model.predict(x_train)
```

Evaluating

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print('Test Accuracy',accuracy_score(y_test,r_y_predict))
print('Train Accuracy',accuracy_score(y_train,r_y_train))
```

```python
pd.crosstab(y_test,r_y_predict)
```

```python
print(classification_report(y_test,r_y_predict))
```
